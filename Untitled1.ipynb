### week1

```python
import pandas as pd
from sklearn.impute import SimpleImputer
def generate_example_dataset():
    data = {
        'PassengerId': [1, 2, 3, 4, 5],
        'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
        'Age': [22, None, 25, None, 30],
        'Fare': [7.25, 71.28, None, 8.05, 10.5],
        'Survived': [0, 1, 1, 0, 1]
    }
    return pd.DataFrame(data)
def clean_dataset(df):
    print("Original Dataset:")
    print(df)
    numeric_df = df.select_dtypes(include='number')
    imputer = SimpleImputer(strategy='mean')
    df_cleaned = pd.DataFrame(imputer.fit_transform(numeric_df), columns=numeric_df.columns)
    print("\nCleaned Dataset:")
    print(df_cleaned)
if __name__ == "__main__":
    example_dataset = generate_example_dataset()
    clean_dataset(example_dataset)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset
data_set = sns.load_dataset('iris')

# Define feature matrix (X) and target vector (y)
x = data_set.iloc[:, :-1].values
y = data_set.iloc[:, 1].values

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

# Create a linear regression model
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Predicting the test set results
y_pred = regressor.predict(x_test)
x_pred = regressor.predict(x_train)

# Visualizing the training set results
plt.scatter(x_train[:, 0], y_train, color="green")  # Assuming the first column for plotting
plt.plot(x_train[:, 0], x_pred, color="red")
plt.title("Sepal Length vs Sepal Width (Training Dataset)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

# Visualizing the test set results
plt.scatter(x_test[:, 0], y_test, color="blue")  # Assuming the first column for plotting
plt.plot(x_train[:, 0], x_pred, color="red")
plt.title("Sepal Length vs Sepal Width (Test Dataset)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```
